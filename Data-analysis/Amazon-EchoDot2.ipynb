```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Dataset
import kagglehub

path= kagglehub.dataset_download('PromptCloudHQ/amazon-echo-dot-2-reviews-dataset')
```

```python
#List CSV
csv_files=[x for x in os.listdir(path) if x.endswith('.csv')]

#Read the files and joind them in a DataFrame
dataframes=[]

for file in csv_files:
    c= os.path.join(path, file)
    df= pd.read_csv(c)
    dataframes.append(df)

df_f= pd.concat(dataframes, ignore_index=True)
```

```python
df_f.drop(['Title','Review Text','Review Date','Review Useful Count','Declaration Text'], axis=1)
```

```python
df0=df_f.fillna(0)
```

Data Visualization

```python
sns.histplot(data=df0['Rating'], color='purple')
plt.title('Rating Count')
plt.show()
```

```python
#Frequency Plot
sns.countplot(data=df0, x='Review Color', color='orange')
plt.title('Frequency by Color')
plt.show()

#Pie Plot
freq_color=df0['Review Color'].value_counts()
plt.figure()
plt.pie(freq_color, labels=freq_color.index,
        autopct='%1.2f%%', startangle=90, colors=plt.cm.Set3.colors   
        )
plt.show()
```

```python
#Frequency Plot
sns.countplot(data=df0, x='Configuration Text', color='green')
plt.title('Frequency by Configuration Text')
plt.show()

#Pie Plot
freq_config= df0['Configuration Text'].value_counts()
plt.figure()
plt.pie(freq_config, labels=freq_config.index,
        autopct='%1.2f%%', startangle=90, colors=plt.cm.Set3.colors   
        )
plt.show()
```

```python
#Frequency Plot
sns.countplot(data=df0, x='User Verified', color='yellow')
plt.title('Frequency by User Verified')
plt.show()

#Pie Plot
freq_user= df0['User Verified'].value_counts()
plt.figure()
plt.pie(freq_user, labels=freq_user.index,
        autopct='%1.2f%%', startangle=90, colors=plt.cm.Set3.colors   
        )
plt.show()
```
